## Derivation for EOM (copied from DeriveEOM.ipynb)

```python
mb = 6.
Jxb = 10.
Jyb = 10.
Jzb = 16.

mw = 1.
Jxw = 0.075
Jyw = 0.075
Jzw = 0.125
lw = 1.1
```

```python
# roll, pitch, yaw angles
phi, theta, psi = sym.symbols('phi, theta, psi')

# angular velocities
w_x, w_y, w_z = sym.symbols('w_x, w_y, w_z')

# torques
tau_1, tau_2, tau_3, tau_4 = sym.symbols('tau_1, tau_2, tau_3, tau_4')

# resultant torques
lt = sym.nsimplify(lw) * sym.sqrt(2) / 2
T1 = - tau_1 * sym.Matrix([[lt], [0], [lt]])
T2 = - tau_2 * sym.Matrix([[-lt], [0], [lt]])
T3 = - tau_3 * sym.Matrix([[0], [lt], [lt]])
T4 = - tau_4 * sym.Matrix([[0], [-lt], [lt]])
T = T1 + T2 + T3 + T4

# parameters
Jx = sym.nsimplify(Jxb + 4 * mw * lw**2)
Jy = sym.nsimplify(Jyb + 4 * mw * lw**2)
Jz = sym.nsimplify(Jzb + 4 * mw * lw**2)

# rotation matrices
Rx = sym.Matrix([[1, 0, 0], [0, sym.cos(phi), -sym.sin(phi)], [0, sym.sin(phi), sym.cos(phi)]])
Ry = sym.Matrix([[sym.cos(theta), 0, sym.sin(theta)], [0, 1, 0], [-sym.sin(theta), 0, sym.cos(theta)]])
Rz = sym.Matrix([[sym.cos(psi), -sym.sin(psi), 0], [sym.sin(psi), sym.cos(psi), 0], [0, 0, 1]])

# angular velocity to angular rates
ex = sym.Matrix([[1], [0], [0]])
ey = sym.Matrix([[0], [1], [0]])
ez = sym.Matrix([[0], [0], [1]])
M = sym.simplify(sym.Matrix.hstack((Ry * Rz).T * ex, Rz.T * ey, ez).inv(), full=True)

# euler's equations
euler = sym.Matrix([[(1 / Jx) * (T[0] + (Jy - Jz) * w_y * w_z)],
                   [(1 / Jy) * (T[1] + (Jz - Jx) * w_z * w_x)],
                   [(1 / Jz) * (T[2] + (Jx - Jy) * w_x * w_y)]])

# equations of motion
f = sym.simplify(sym.Matrix.vstack(M * sym.Matrix([[w_x], [w_y], [w_z]]), euler), full=True)
```

The equations of motion have this form:

$$\begin{bmatrix} \dot{\phi} \\ \dot{\theta} \\ \dot{\psi} \\ \dot{w_x} \\ \dot{w_y} \\ \dot{w_z} \end{bmatrix} = f\left(\phi, \theta, \psi, w_x, w_y, w_z, \tau_1, \tau_2, \tau_3, \tau_4\right)$$

Here is the function $f$:

## Sensor model

```python
alpha, delta = sym.symbols('alpha, delta')#right ascension of the star;
                                          #declination of the start
# Scope radius
r = 0.8 / 2.1
# Position of star in space frame
p_star_in_space = sym.Matrix([[sym.cos(alpha) * sym.cos(delta)],
                              [sym.sin(alpha) * sym.cos(delta)],
                              [sym.sin(delta)]])

# Orientation of body frame in space frame
R_body_in_space = Rx * Ry * Rz

# Position of star in body frame (assuming origin of body and space frames are the same)
p_star_in_body = R_body_in_space.T * p_star_in_space

# Position of star in image frame
p_star_in_image = (1 / sym.nsimplify(r)) * sym.Matrix([[p_star_in_body[1] / p_star_in_body[0]],
                                                       [p_star_in_body[2] / p_star_in_body[0]]])

# Sensor model for each star
g = sym.simplify(p_star_in_image, full=True)
```

The sensor model has this form for each star:

$$\zeta = g(\phi, \theta, \psi, \alpha, \delta)$$

## State-space representation

```python
#define state vector X
phi_e,theta_e,psi_e,w_x_e,w_y_e,w_z_e,tau_1_e,tau_2_e, \
tau_3_e,tau_4_e = sym.symbols('phi_e,theta_e,psi_e,w_x_e,\
w_y_e,w_z_e,tau_1_e,tau_2_e,tau_3_e,tau_4_e') 
#defining symbolic expressions for equilibrium values

x = sym.Matrix([[phi-phi_e],[theta-theta_e],[psi-psi_e],[w_x-w_x_e],
                [w_y-w_y_e],[w_z-w_z_e]])

x #state vector
```

```python
#defining input vector
u = sym.Matrix([[tau_1-tau_1_e],[tau_2-tau_2_e],[tau_3-tau_3_e],[tau_4-tau_4_e]])
u
```

```python
#obtain matrix C based on output g (sensor model)
C1_num = sym.lambdify((phi,theta,psi,alpha,delta),g.jacobian([phi,theta,psi,w_x,w_y,w_z]))
alpha_e_1 = 0 #choose values for alpha,delta is equivalent to subsituting values
delta_e_1 = 0 #into g; this is not meant to be equilibiurm values at all
phi_e = 0. #setting equilibrium values
theta_e = 0.
psi_e = 0.

C1 = C1_num(phi_e,theta_e,psi_e,alpha_e_1,delta_e_1)
# we need to stack 3 output matrices together because 
#we have 3 stars by default
```

```python
C2_num = sym.lambdify((phi,theta,psi,
alpha,delta),g.jacobian([phi,theta,psi,w_x,w_y,w_z]))

alpha_e_2 = 0.15
delta_e_2 = 0
C2 = C2_num(phi_e,theta_e,psi_e,alpha_e_2,delta_e_2)
```

```python
C3_num = sym.lambdify((phi,theta,psi,\
alpha,delta),g.jacobian([phi,theta,psi,w_x,w_y,w_z]))

alpha_e_3 = 0
delta_e_3 = 0.15
C3 = C3_num(phi_e,theta_e,psi_e,\
            alpha_e_3,delta_e_3)
```

```python
C_combined = np.block([[C1],[C2],[C3]])
C_combined.shape
```

```python
A_num = sym.lambdify((phi,theta,psi,w_x,w_y,w_z,tau_1,tau_2,tau_3,tau_4),\
                 f.jacobian([phi,theta,psi,w_x,w_y,w_z]))
B_num = sym.lambdify((phi,theta,psi,w_x,w_y,w_z,tau_1,tau_2,tau_3,tau_4),\
                 f.jacobian([tau_1,tau_2,tau_3,tau_4]))

phi_e = 0. #setting equilibrium values
theta_e = 0.
psi_e = 0.
w_x_e = 0
w_y_e = 0
w_z_e = 0
tau_1_e = 0
tau_2_e = 0
tau_3_e = 0
tau_4_e = 0

A = A_num(phi_e,theta_e,psi_e,w_x_e,w_y_e,w_z_e,tau_1_e,\
          tau_2_e,tau_3_e,tau_4_e).astype(float)
B = B_num(phi_e,theta_e,psi_e,w_x_e,w_y_e,w_z_e,tau_1_e,\
          tau_2_e,tau_3_e,tau_4_e).astype(float)
```

```python
#test observability-controllability
#constructing the controllability matrix W
W = B   
for i in range(1, A.shape[0]):
    col = np.linalg.matrix_power(A, i) @ B
    W = np.block([W, col])  #[B,AB,A^2@B,A^3@B.....]

print(np.linalg.matrix_rank(W)) #rank of W
print(A.shape[0]) #number of states in A
```

```python
#constructing the observability matrix O
O=C_combined
for i in range(1,A.shape[0]):
    col = C_combined@np.linalg.matrix_power(A,i)
    O = np.block([[O],[col]])
print(np.linalg.matrix_rank(O))
print(A.shape[0])
```

```python
def LQR_contr(A,B,Q,R): #LQR for controller
    x0 =np.array([[0],[0],[0],[0],[0],[0]])
    P = linalg.solve_continuous_are(A, B, Q, R) #finding optimal cost matrix
    K = linalg.inv(R) @  B.T @ P #gain matrix 
    u = -K@x0
    total_cost = (x0.T @ P @ x0).item()

    return P,K,u,total_cost
```

```python
def LQR_obsv(A,C,R,Q):  #LQR for observer
    P = linalg.solve_continuous_are(A.T,C.T,np.linalg.inv(R), \
                  np.linalg.inv(Q)) #finding optimal cost matrix
    L = P@C.T@Q #injection matrix 
    return L
```

```python
Q = 1*np.eye(A.shape[0]) 
R = 1*np.eye(C_combined.shape[1])
L= LQR_obsv(A,C_combined,R,Q)
```

```python
print(linalg.eigvals(A-L@C_combined)) #real parts are all negative! nice
```

```python
#now let's find gain matrix K for the controller
Q = 50*np.eye(A.shape[0])
R = 1*np.eye(B.shape[1])
P,K,u,total_cost = LQR_contr(A,B,Q,R)
K
```

```python
print(linalg.eigvals(A-B@K))#real parts are once again all negative!
```

Controllability-observability is satisfied.However, since the exact state parameters are not returned by the sensor, we can only approximate the state variables using a set of initial conditions and advance this IVP in time

```python
global g_stack
g_stack = np.block([[g.subs([(phi,0.),(theta,0.),(psi,0.),\
            (alpha,0),(delta,0)])],        #1st star
            [g.subs([(phi,0.),(theta,0.),(psi,0.),
            (alpha,0.15),(delta,0.)])],   #2nd star
            [g.subs([(phi,0.),(theta,0.),(psi,0.), #3rd star
            (alpha,0.),(delta,0.15)])]]).astype(float)
#equilibrium values for output


```

```python
class RobotController:

    def __init__(self,dt,phi_e,theta_e,psi_e, tau_1_e,tau_2_e,tau_3_e,tau_4_e,
                 p_e,A, B, C, K, L): #adding disturbances 
        
        self.dt =  dt
        
        self.phi_e = phi_e#setting equilibrium values
        self.theta_e = theta_e
        self.psi_e = psi_e
        self.w_x_e = w_x_e
        self.w_y_e = w_y_e
        self.w_z_e = w_z_e
        self.tau_1_e = tau_1_e
        self.tau_2_e = tau_2_e
        self.tau_3_e = tau_3_e
        self.tau_4_e = tau_4_e
        self.p_e = p_e
        
        # Define state-space model
        self.A = A
        self.B = B
        self.C =C
        # Define controller and observer gains
        self.K = K
        self.L = L
        # Run the reset function to initialize the state estimate
        self.reset()
        
    
    def reset(self):
        # Initialize the state estimates
        self.xhat = np.zeros((6,1))
    
    def run(self,p,v):
        
        if np.isnan(p).any() ==True:
            p = self.p_previous #this ensures that the positions of the stars are
                                # recorded until they vanish from sight
                                # note: this is referenced from 'tylerbr2'
                        
        # Apply controller
        u = -self.K@self.xhat 
        # Convert input to actuator commands
        tau = np.array([u[0,0]+self.tau_1_e, u[1,0]+self.tau_2_e, u[2,0]\
                        +self.tau_3_e, u[3,0]+self.tau_4_e])
         
        # Define output
        y = np.array(p-self.p_e)
        y = np.nan_to_num(y).astype(float)
        self.xhat+=self.dt * (self.A @ self.xhat + self.B @ u -self.L @ (self.C @ self.xhat - y.T))
        self.p_previous = p
        
        return tau
    
```

```python
robot = ae353_spacecraft.RobotSimulator(display=False)
```

```python
controller=RobotController(dt=robot.dt,phi_e=phi_e,
                  theta_e=theta_e,
                  psi_e=psi_e,
                  tau_1_e=tau_1_e,
                  tau_2_e = tau_2_e,
                  tau_3_e = tau_3_e,
                  tau_4_e=tau_4_e,
                  p_e=g_stack.T,
                  A=A,
                  B=B,
                  C=C_combined,
                  K=K,
                  L=L)
```

### Collecting aggregate statistics of error in roll,pitch, and yawl angles

```python
phi_errs = []
theta_errs = []
psi_errs = []
settling_times = []

num_trials = 100
for trial in range(num_trials):
    robot.reset()
    controller.reset()
    
    run_time = 10.
    num_steps = int(run_time/robot.dt)
    data = {
    't': np.empty(num_steps, dtype=float),
    'v': np.empty((4, num_steps), dtype=float),
    'p': np.empty((2 * len(robot.stars), num_steps), dtype=float),
    'tau': np.empty((4, num_steps), dtype=float),
    'rpy': np.empty((3, num_steps), dtype=float),
    'angvel': np.empty((3, num_steps), dtype=float),
    'xhat': np.empty((6, num_steps), dtype=float),
    'x' : np.empty((6,num_steps),dtype=float) }
    
    start_time = time.time()
    for step in range(num_steps):
    # Get the current time
        t = robot.dt * step

        # Get the sensor measurements
        p, v = robot.get_sensor_measurements()
        
        # Choose the actuator command (by running the controller)
        tau= controller.run(p,v)
        if tau.shape != (4, ):
            raise Exception(f'invalid desired torques: {tau}')

        # Get orientation and angular velocity as ground truth
        rpy, angvel = robot.get_rpy_and_angvel()

        # Log the data from this time step
        data['t'][step] = t
        data['v'][:, step] = v
        data['p'][:, step] = p
        data['tau'][:, step] = tau
        data['rpy'][:, step] = rpy
        data['angvel'][:, step] = angvel
        data['xhat'][:, step] =controller.xhat.flatten()
        #print(data['xhat'])

        # Send the actuator commands to robot and go forward one time step
        robot.set_actuator_commands(tau)
        robot.step(t=(start_time + (robot.dt * (step + 1))))
    
    phi_est = data['xhat'][0, :] + controller.phi_e 
    theta_est = data['xhat'][1, :] + controller.theta_e
    psi_est = data['xhat'][2, :] + controller.psi_e
    w_x_est = data['xhat'][3, :] + controller.w_x_e
    w_y_est = data['xhat'][4, :] + controller.w_y_e
    w_z_est = data['xhat'][5, :] + controller.w_z_e
    
    
    #filtering out terms that are very close to or equal to 0
    
    for i in range(len(data['rpy']+1)):
        phi_err = np.sum((data['rpy'][0,i]-phi_est)**2) #sum sqaured error
        theta_err = np.sum((data['rpy'][1,i]-theta_est)**2)
        psi_err = np.sum((data['rpy'][2,i]-psi_est)**2)
        
            
    phi_errs.append(phi_err/(len(data['t']))) #divide by numer of 
                                    #measurements to get mean squared error
    theta_errs.append(theta_err/(len(data['t'])))
    psi_errs.append(psi_err/(len(data['t'])))
    
    if (trial + 1) % 10 == 0:
        print(f' {trial + 1} / {num_trials}')
    
```

### Analysis question: How do your results change with different amounts of sensor noise?

```python
robot.reset()
controller.reset()
run_time = 10
num_steps = int(run_time/robot.dt)
#print(num_steps)
# Create a dictionary in which to store results
data = {
    't': np.empty(num_steps, dtype=float),
    'v': np.empty((4, num_steps), dtype=float),
    'p': np.empty((2 * len(robot.stars), num_steps), dtype=float),
    'tau': np.empty((4, num_steps), dtype=float),
    'rpy': np.empty((3, num_steps), dtype=float),
    'rpy_mes': np.empty((3, num_steps), dtype=float),
    'angvel': np.empty((3, num_steps), dtype=float),
    'xhat': np.empty((6, num_steps), dtype=float),
    'x' : np.empty((6,num_steps),dtype=float) ,
}

# Run the simulation loop
start_time = time.time()
for step in range(num_steps):
    # Get the current time
    t = robot.dt * step
    
    # Get the sensor measurements
    p, v = robot.get_sensor_measurements()
   
        
    tau= controller.run(p,v)
    if tau.shape != (4, ):
        raise Exception(f'invalid desired torques: {tau}')
    
    rpy, angvel = robot.get_rpy_and_angvel()
    rpy_mes = (rpy[0] + 0.2*np.random.rand(),rpy[1] + \
               0.2*np.random.rand(),rpy[2] + \
               0.2*np.random.rand()) 
    #adding disturbances to rpy(a tuple)
    
    data['t'][step] = t
    data['v'][:, step] = v
    data['p'][:, step] = p
    data['tau'][:, step] = tau
    data['rpy'][:, step] = rpy
    data['rpy_mes'][:, step] = rpy_mes
    data['angvel'][:, step] = angvel
    data['xhat'][:, step] =controller.xhat.flatten()
    robot.set_actuator_commands(tau)
    robot.step(t=(start_time + (robot.dt * (step + 1))))

#updating estimations
phi_est = data['xhat'][0, :] + controller.phi_e 
theta_est = data['xhat'][1, :] + controller.theta_e
psi_est = data['xhat'][2, :] + controller.psi_e
w_x_est = data['xhat'][3, :] + controller.w_x_e
w_y_est = data['xhat'][4, :] + controller.w_y_e
w_z_est = data['xhat'][5, :] + controller.w_z_e
```
